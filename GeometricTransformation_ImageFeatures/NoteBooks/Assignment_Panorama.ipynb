## image stitching pipeline



> This figure illustrates the `stitching module pipeline` implemented in the `Stitcher class`. Using that class it's possible to configure/remove some steps, i.e. adjust the stitching pipeline according to the particular needs. All building blocks from the pipeline are available in the detail namespace, one can combine and use them separately.

```python
images = []

dirName = "scene"

imagefiles = [DATA_PATH + "images/{}/".format(dirName) + f for f in os.listdir(DATA_PATH+"images/"+dirName) if f.endswith(".jpg")]
imagefiles.sort()

destination = "{}_result.png".format(dirName)
plt.figure(figsize=[20,15])
i=1
for filename in imagefiles:
    img = cv2.imread(filename)
    images.append(img)

###
stitcher = cv2.Stitcher_create()
status, result = stitcher.stitch(images)

# Check stitiching status
if status == cv2.STITCHER_OK:
    cv2.imwrite(destination, result)
    plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
    plt.show()
else:
    print("Image stitching failed with status:", status)
###
```
